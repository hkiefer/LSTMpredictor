```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#import datetime as dt
from wwo_hist import retrieve_hist_data 
from LSTMpredictor import *
from loaddata import *
```

## Load Dataset from WWO (key needed)

```python
trj = pd.read_csv("Berlin.csv", index_col = "date_time")

trj.index = pd.to_datetime(trj.index)
trj = trj.resample("7D").mean()
trj = trj.fillna(method = 'ffill')

trj['maxtempC'] = trj['maxtempC'].astype('float')

plt.plot(trj['maxtempC'].values)
#plt.yticks(np.arange(np.min(trj['maxtempC']), np.max(trj['maxtempC']), 1))
trj.head()
```

## Build and Fit LSTM

```python
lp = LSTMpredictor(interval = "weekly", n_prev = 100, cut = int(len(trj)*0.8), verbose_plot = True)
filename = 'weather.h5' #we can load the already trained model
model = lp.build_fit_LSTM(trj = trj, value = 'maxtempC', hidden_layers = 2, units = 16, dropout = 0.5, epochs = 100, batch_size = 256, filename = filename)
```

## Predict future trajectory

```python
#perform a multi-step prediction with moving window
filename = 'weather.h5'
lp = LSTMpredictor(interval = "weekly", n_prev = 100, verbose_plot = True)
pred_trj = lp.moving_test_window_preds(trj, value = 'maxtempC', n_steps = 60, filename = filename, start = len(trj)-60)
```
