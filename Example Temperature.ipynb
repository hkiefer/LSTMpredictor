```python
trj = pd.read_csv("Berlin.csv", index_col = "date_time")

trj.index = pd.to_datetime(trj.index)
trj = trj.resample("7D").mean()
trj = trj.fillna(method = 'ffill')

trj['maxtempC'] = trj['maxtempC'].astype('float')

plt.plot(trj['maxtempC'].values)
#plt.yticks(np.arange(np.min(trj['maxtempC']), np.max(trj['maxtempC']), 1))
trj.head()
```

## Build and Fit LSTM

```python
lp = LSTMpredictor(interval = "weekly", n_prev = 100, cut = int(len(trj)*0.8), verbose_plot = True)
filename = 'weather.h5'
model = lp.build_fit_LSTM(trj = trj, value = 'maxtempC', hidden_layers = 2, units = 16, dropout = 0.5, epochs = 10, batch_size = 256, filename = filename)
```
